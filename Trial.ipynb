```python
# find the top freq for each note, lets say higher than 10e7
threshold = 10e6
top_freq_A2 = []
c_A2 = abs(c_A2)
for i, magn in enumerate(c_A2):
    freq = np.linspace(0,fs, len(c_A2))[i]
    if magn >= threshold:
        freq = np.linspace(0,fs, len(c_A2))[i]
        if (len(top_freq_A2) == 0) or (abs(freq - top_freq_A2[-1])>20):
            top_freq_A2.append(freq)
            
print(top_freq_A2)

threshold = 10e6
top_freq_B2 = []
c_B2 = abs(c_B2)
for i, magn in enumerate(c_B2):
    freq = np.linspace(0,fs, len(c_B2))[i]
    if magn >= threshold:
        freq = np.linspace(0,fs, len(c_B2))[i]
        if (len(top_freq_B2) == 0) or (abs(freq - top_freq_B2[-1])>20):
            top_freq_B2.append(freq)
            
print(top_freq_B2 ) 
```
