```python
import matplotlib.pyplot as plt
from scipy.fftpack import rfft, ifft
from scipy.io import wavfile
import numpy as np
```

```python
instruments = ['AcousticGuitar', 'CathedralOrgan', 'Clarinet', 'FrenchHorn', 'GrandPiano', 'Harp',
              'Koto', 'RockGuitar', 'Saxophone', 'Guitar&FrenchHorn', 'GrandPiano&Clarinet']


# sampling a representative time frame for each instrument
samples = dict()

# saving the frequencies
freqs = dict()
fig, axs = plt.subplots(1,2,figsize=(20,10))

instrument = instruments[0]
print(instrument)
fs, data = wavfile.read('sounds/'+instrument+'.wav') # load the data
freqs[instrument] = fs

# sample selected = 1 to 2 sec
start = 0*fs
end = 4*fs
samples[instrument] = data.T[0][start:end]
t_length = data.shape[0]

# sample selected = 0.7 to 2 sec
start = int(0.72*fs)
end = int(0.74*fs)
samples[instrument] = data.T[0][start:end]
t_length = data.shape[0]
# plt.plot(np.linspace(0,t_length/fs, t_length), data.T[0], label = 'Entire recording')
axs[0].plot(np.linspace(start/fs, end/fs, (end-start)), samples[instrument])
axs[0].set(xlabel='t',title=instrument)
# plt.ylabel('Magnitude', fontsize = 15)



instrument = instruments[3]
print(instrument)
fs, data = wavfile.read('sounds/'+instrument+'.wav') # load the data
freqs[instrument] = fs

# sample selected = 1 to 2 sec
start = 0*fs
end = 4*fs
samples[instrument] = data.T[0][start:end]
t_length = data.shape[0]

# sample selected = 0.7 to 2 sec
start = int(0.73*fs)
end = int(0.75*fs)
samples[instrument] = data.T[0][start:end]
t_length = data.shape[0]
# plt.plot(np.linspace(0,t_length/fs, t_length), data.T[0], label = 'Entire recording')
# plt.plot(np.linspace(0,t_length/fs, t_length), data.T[0], label = 'Entire recording')
axs[1].plot(np.linspace(start/fs, end/fs, (end-start)), samples[instrument])
axs[1].set(xlabel='t',title=instrument)
# plt.ylabel('Magnitude', fontsize = 15)
```
